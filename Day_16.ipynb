**column removal: license**
- The license column contained over 99% missing values, making it non-informative.
- it is being removed to reduce noise and improve data quality for downstream analysis

```python
df = df.drop(columns=['license'])
```

```python
df.columns
```

```python
drop_cols = [
    "house_rules",      # unstructured text
    "country_code",     # constant
    "country",          # constant
]
df = df.drop(columns=drop_cols, errors="ignore")
```

```python
for col in ["price", "service fee"]:
    df[col] = df[col].replace(r"[\$,]", "", regex=True).astype(float)
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df["last review"] = pd.to_datetime(df["last review"], errors="coerce")
```

```python
df["reviews per month"] = df["reviews per month"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["NAME"] = df["NAME"].fillna("Unknown")
df["host name"] = df["host name"].fillna("Unknown")
```

```python
df["host_identity_verified"] = df["host_identity_verified"].fillna("Unknown")
```

```python
df["neighbourhood"] = df["neighbourhood"].fillna("Unknown")
df["neighbourhood group"] = df["neighbourhood group"].fillna("Unknown")
```

```python
df = df.dropna(subset=["lat", "long"])
```

```python
df["minimum nights"] = df["minimum nights"].fillna(
    df["minimum nights"].median()
)
```

```python
df["number of reviews"] = df["number of reviews"].fillna(0)
```

```python
df["review rate number"] = df["review rate number"].fillna(
    df["review rate number"].median()
)
```

```python
df["calculated host listings count"] = df[
    "calculated host listings count"
].fillna(df["calculated host listings count"].median())
```

```python
df["availability 365"] = df["availability 365"].fillna(
    df["availability 365"].median()
)
```

```python
df["last review"] = pd.to_datetime(df["last review"], errors="coerce")
df["has_reviews"] = df["last review"].notna().astype(int)
```

```python
df.isna().sum()
```

```python
df["country code"] = df["country code"].fillna("Unknown")
```

```python
df["instant_bookable"] = df["instant_bookable"].fillna("Unknown")
```

```python
df["cancellation_policy"] = df["cancellation_policy"].fillna("Unknown")
```

```python
df["Construction year"] = df["Construction year"].fillna(
    df["Construction year"].median()
)
```

```python
df.isna().sum()
```

Today’s work focused on turning a raw, messy real-world dataset into a model-ready asset, using decisions aligned with industry data science and AI engineering standards.

We began by auditing missing values column-by-column, instead of blindly filling or dropping data. Each feature was evaluated based on its business meaning, not just data type. For example, last review was intentionally left missing because absence of reviews is meaningful information, not an error. This logic was strengthened by creating a derived feature (has_reviews) to explicitly capture that signal.

Categorical fields such as country code, instant_bookable, and cancellation_policy were filled with "Unknown" rather than forcing incorrect assumptions. This preserves data integrity and reflects real production scenarios where information may be unavailable or withheld.

For numerical but non-continuous fields like Construction year, median imputation was applied to avoid skew while maintaining dataset size. This choice balances statistical soundness with practical modeling needs.

Throughout the process, we avoided data leakage, respected domain context, and made decisions that keep the dataset flexible for EDA, ML modeling, and future feature engineering.

By the end of the day, the dataset reached a clean, explainable, and production-ready state, suitable for:

Regression or classification models

Feature importance analysis

Business-driven insights
