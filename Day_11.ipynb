**Feature Intent Table**
-----------------------------------------------------------------------------
Column	      Why it exists	           Business question it answers
-----------------------------------------------------------------------------
Year	      Temporal grouping   	   How climate metrics change over time
Season	      Climate segmentation	   Seasonal risk & behavior patterns
Temp Max      Heat intensity	       Are heatwaves becoming more severe?
Temp_Range	  Variability	           Is weather becoming more unstable?
Rain	      Precipitation volume	   Are rainfall patterns extreme or skewed?

```python
df.groupby("Year")["Temp Max"].mean()
```

**Yearly Maximum Temperature Trend**

Purpose:
- To validate long-term warming patterns after cleaning and feature selection.

Insight:
- A gradual upward drift indicates climate change signals rather than short-term noise.
- Supports earlier heatwave analysis (Day 10).

Why this matters:
- Rising max temperatures increase heatwave frequency, energy demand, and health risk.


```python
df.groupby("Year")["Rain"].mean()
```

**Yearly Average Rainfall Analysis**

Purpose:
- To understand long-term rainfall behavior across decades.

Insight:
- Rainfall averages are volatile due to extreme events.
- Mean alone is not a reliable indicator of flood or drought risk.

Connection:
- Justifies why extreme rainfall thresholds were introduced in Day 10.


```python
df["Temp_Range"].describe()
```

**Daily Temperature Range Distribution**

Purpose:
- To quantify daily climate volatility rather than averages.

Insight:
- Wide temperature ranges stress crops, ecosystems, and infrastructure.
- Volatility is often more damaging than gradual warming.

Business relevance:
- Useful for agriculture planning and climate resilience studies.


```python
df.groupby("Year")["Temp Max"].mean().plot(
    title="Yearly Average Maximum Temperature Trend"
)
```

**Yearly Average Maximum Temperature Trend**

Observation:
- A steady rise in average maximum temperature is visible across decades.
- Post-2000 years show accelerated warming with higher peaks.

Interpretation:
- Indicates long-term climate warming rather than short-term variability.
- Aligns with increased heatwave frequency observed in Day 10.

Why this matters:
- Rising maximum temperatures increase health risks, crop stress, and energy demand.
- Serves as a foundational trend supporting extreme climate event analysis.
