```python
climate_df['Date'] = climate_df['Date'].astype(str)
# this avoids silent type issues
```

```python
climate_df.dtypes
```

```python
excel_mask = climate_df['Date'].str.isnumeric()

climate_df.loc[excel_mask, 'Date'] = pd.to_datetime(
    climate_df.loc[excel_mask, 'Date'].astype(int),
    unit='D',
    origin=1899-12-30
)
```

```python
date_mask = excel_mask

climate_df.loc[date_mask, 'Date'] = pd.to_datetime(
    climate_df.loc[date_mask, 'Date'],
    dayfirst=True,
    errors='coerce'
)
```

```python
climate_df['Date'] = pd.to_datetime(
    climate_df['Date'],
    errors='coerce'
)
```
